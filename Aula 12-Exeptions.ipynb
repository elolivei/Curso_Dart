### TRATAMENTO DE EXCEÇÕES NO DART
- Uma exceção é um erro que ocorre dentro do programa. 
- Quando ocorre uma exceção dentro de um programa, o fluxo normal do programa é interrompido e termina de forma anormal, exibindo o erro e a pilha de exceções como saída. 
- Portanto, uma exceção deve ser observada para evitar que o aplicativo seja encerrado

* Exceções integradas no Dart: 
*


* Cada exceção embutida no Drat vem em uma classe predefinida chamada ***Exception***.
* Para evitar a exceção do programa, usamos os blocos ***try / on / catch*** no Dart.



```python
void main() {
  int geek = 10 ;
  int geek2 = 0;
  try{
    var geek3 = geek ~/ geek2;
    print(geek2);
  }
  on UnsupportedError{
    print("Erro divisão por zero.");
  }
}
```

### Prática
* Escreva o código de uma calculadora onde será digitado os valores e a operação.
* Não esqueça de prever a divisão por zero.
