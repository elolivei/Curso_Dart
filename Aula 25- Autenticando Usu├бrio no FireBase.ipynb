# Autenticação de usuário.
* Controlar quem pode usar a aplicação é uma necessidade de boa parte das aplicações comenrciais.
* Vamos aprender como autenticar o usuario usando o Firebase.

## Já existe uma api pronta para essa finalidade?
* SIM! Vamos fazer o passo a passo do codlab abaixo para aprender. 

https://firebase.google.com/codelabs/firebase-auth-in-flutter-apps?hl=pt-br#0


### Como nem tudo são flores, a api flutterfire_ui foi depreciada.
* Então vamos atualizar a aplicação com os novos widgets.

https://pub.dev/packages/flutterfire_ui

Vamos usar : firebase_ui_auth
