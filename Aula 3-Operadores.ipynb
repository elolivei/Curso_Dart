### Operador em dart 
* é um **símbolo** que é usado para **executar operações**. Por exemplo: +, -, *, / etc.

Existem muitos tipos de operadores em dart que são fornecidos abaixo:

* Operadores aritméticos
* Operadores relacionais
* Operadores de teste de tipo
* Operadores bit a bit
* Operadores de atribuição
* Operadores lógicos
* Operador Condicional
* Operador de notação em cascata

### operadores aritimeticos 


```python
void main()
{
    int a = 2;
    int b = 3;
 
    // Somando
    var c = a + b;
    print("A Soma de a e b é $c");
 
    // Subtraindo
    var d = a - b;
    print("A diferença entre a e b é $d");
 
    // Inversão de sinal
    var e = -d;
    print("Invertendo o sinal de d temos $e");
 
    // Multiplicando
    var f = a * b;
    print("O produto de a e b é $f");
 
    // Divisão
    var g = b / a;
    print("Dividindo a por b temos $g");
 
    // Forçando uma divisão retornar inteiro
    var h = b ~/ a;
    print("A parte inteira da divisão de a por b é $h");
 
    // O resto de uma divisão
    var i = b % a;
    print("O resto da divisão de a por b é $i");
}
```

###   Operadores relacionais:
* Essa classe de operadores contém aqueles operadores que são usados ​​para realizar operações relacionais nos operandos. 



```python
void main()
{
    int a = 2;
    int b = 3;
 
    // Greater between a and b
    var c = a > b;
    print("a is greater than b is $c");
 
    // Smaller between a and b
    var d = a < b;
    print("a is smaller than b is $d");
 
    // Greater than or equal to between a and b
    var e = a >= b;
    print("a is greater than b is $e");
 
    // Less than or equal to between a and b
    var f = a <= b;
    print("a is smaller than b is $f");
 
    // Equality between a and b
    var g = b == a;
    print("a and b are equal is $g");
 
    // Unequality between a and b
    var h = b != a;
    print("a and b are not equal is $h");
}
```

## Operador de atribuição
Atribuindo um valor a uma variável



```python
void main() {

var num1 = 5; // 5
print("Após Atribuição: $num1");
num1 -= 4; // num1 = num1 - 4
print("Após Atribuição com subtração: $num1");
num1 +=4; // num1 = num1 + 4
print("Após Atribuição com soma: $num1");
num1 ~/= 5; // num1 = num1 ~/ 5
print("Após Atribuição com divisão: $num1");
num1 *= 5; // num1 = num * 5
print("Após Atribuição com multiplicação: $num1");
}
```
