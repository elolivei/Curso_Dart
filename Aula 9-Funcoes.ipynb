### Funções
* Em Dart, a estrutura de uma função é parecida com as de outras linguagens.
* Uma função é um grupo de instruções que, juntas, executam uma tarefa específica.
* Sua função pode ou não retornar um valor/objeto 
* Colocando o tipo de retorno como void ou não colocando nada antes da declaração da função o retorno será vazio.

```python
void main(){
  printIntro();
  calcSoma(10.0, 15.0);
  print(calcMult(10.0, 15.0));
}

// Função sem parametro e sem retorno
printIntro(){
  print("Seja bem-vindo");
}

// aqui estão os parâmetros, você pode colocar quantos parâmentros você quiser dentro de uma função
// sem retorno
void calcSoma(double a, double b){
  double res = a + b;
  print(res);
}

// Função com parametro e retorno
double calcMult(double a, double b){
  double res = a * b;
  // a função irá retornar um valor, e como o valor retornado é double, tem que retirar o void e colocar double no seu lugar, assim é necessário dar um return
  return res;
}
```

* Dart é uma linguagem orientada a objetos, então até mesmo funções são objetos e temos um tipo, Function.
* Isso significa que as funções podem ser atribuídas a variáveis ou passadas como argumentos para outras funções. 
* Você também pode chamar uma instância de uma classe Dart como se fosse uma função.
* Para funções que contêm apenas uma expressão, você pode usar uma abreviação sintaxe: ***=>***


```python
var _nobleGases = {
  2: 'helium',
  10: 'neon',
  18: 'argon',
};

bool isNoble(int atomicNumber) => _nobleGases[atomicNumber] != null;

void main(){

  var nAton1 = 10;
  var nAton2 = 12;

  if(isNoble(nAton1))
    print("$nAton1 é um gás nobre");
  else
    print("$nAton1 não é um gás nobre");

  if(isNoble(nAton2))
    print("$nAton2 é um gás nobre");
  else
    print("$nAton2 não é um gás nobre");

}
```

### Parâmetros
* O Dart suporta parâmetros opcionais nomeados e parâmetros opcionais posicionais
* Os parâmetros opcionais do Dart são opcionais , pois o chamador não precisa especificar um valor para o parâmetro ao chamar a função.
* Parâmetros opcionais só podem ser declarados após quaisquer parâmetros necessários.
* Os parâmetros opcionais podem ter um valor padrão, usado quando um chamador não especifica um valor.

### Parâmetros opcionais posicionais
* Um parâmetro agrupado por [ ]é um parâmetro opcional posicional
* Você pode especificar vários parâmetros posicionais para uma função

```python
void main(){
  
  print ("Soma dois elementos ${calcSoma(10.0, 15.0)}");
  print ("Soma trez elementos ${calcSoma(10.0, 15.0, 20.0)}");
}

calcSoma(double a, double b, [double? c]) {
  if (c == null) 
    return(a + b);
  else 
    return (a+b+c);
  
}
```

### Parâmetros opcionais nomeados
* Um parâmetro agrupado por { }é um parâmetro opcional nomeado.
* Você deve usar o nome do parâmetro ao chamar a função.
* Você pode especificar vários parâmetros nomeados para uma função.
* Vamos alterar a função para parametros nomeados.

```python
void main(){
  
  print ("Soma dois elementos ${calcSoma(10.0, 15.0)}");
  print ("Soma trez elementos ${calcSoma(10.0, 15.0, c:20.0)}");
}

calcSoma(double a, double b, {double? c}) {
  if (c == null) 
    return(a + b);
  else 
    return (a+b+c);
  
}

```

### Definindo valor default para parametro opcional

```python

void main(){
  
  print ("Soma dois elementos ${calcSoma(10.0, 15.0)}");
  print ("Soma trez elementos ${calcSoma(c:20.0 ,10.0, 15.0)}");
}

calcSoma(double a, double b, {double c = 0}) => a+b+c;
```
