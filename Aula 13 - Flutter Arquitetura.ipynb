# Flutter widgets

* O conceito central do framework Flutter é ***In Flutter, Everything is a widget***.
* Widgets são basicamente componentes de interface usados para criar a interface do usuário do aplicativo.
* No Flutter, o aplicativo é em si um widget. 
* O aplicativo é o widget de nível superior e sua interface do usuário é construída usando um ou mais filhos (widgets), que novamente são construídos usando seus widgets filhos. 
* Esse recurso de composabilidade nos ajuda a criar uma interface de usuário de qualquer complexidade.

* Por exemplo, a hierarquia de widgets do aplicativo hello world 


* ***MyApp*** é o widget criado pelo usuário e é construído usando o widget nativo Flutter, ***MaterialApp***.
***MaterialApp*** tem uma propriedade home para especificar a interface do usuário da home page, que é novamente um widget criado pelo usuário, ***MyHomePage***.

* ***MyHomePage*** é construído usando outro widget nativo flutter, ***Scaffold*** que tem duas propriedades – ***body*** e ***appBar***

***body*** é usado para especificar sua interface de usuário principal e ***appBar*** é usado para especificar sua interface de usuário de cabeçalho

O cabeçalho usa o widget nativo flutter, a ***AppBar*** e a interface do usuário do corpo são criadas usando o widget ***Central***.

O widget ***Central*** tem uma propriedade, ***Child***, que se refere ao conteúdo real e é construído usando o widget ***Texto***

```python
import 'package:flutter/material.dart';

void main() => runApp(MyApp());

class MyApp extends StatelessWidget {
   // This widget is the root of your application.
   @override
   Widget build(BuildContext context) {
      return MaterialApp(
         title: 'Hello World Demo Application',
         theme: ThemeData(
            primarySwatch: Colors.blue,
         ),
         home: MyHomePage(title: 'Home page'),
      );
   }
} //MyApp

class MyHomePage extends StatelessWidget {
   MyHomePage({Key? key,required this.title}) : super(key: key);
   final String title;

   @override
   Widget build(BuildContext context) {
      return Scaffold(
         appBar: AppBar(
            title: Text(this.title),
         ),
         body: Center(
            child:
            Text(
               'Hello World',
            ) //Text
         ), //Center
      );//Scaffold
   }
} //MyHomePage
```

### Gestures (Gestos)

* Os widgets Flutter suportam a interação através de um widget especial, o ***GestureDetector***. 
* ***GestureDetector*** é um widget invisível com a capacidade de capturar interações do usuário, como tocar, arrastar, etc., de seu widget filho. 
* Muitos widgets nativos do Flutter suportam a interação através do uso do ***GestureDetector***. Também podemos incorporar o recurso interativo no widget existente, compondo-o com o widget ***GestureDetector***

### Conceito de Estado
* Os widgets Flutter suportam a manutenção do Estado fornecendo um widget especial, ***StatefulWidget***. 
* O widget precisa ser derivado do widget StatefulWidget para suportar a manutenção do estado e todos os outros widgets devem ser derivados do ***StatefulWidget***. 
* Os widgets Flutter são reativos. O widget ***StatefulWidget*** será re-renderizado automaticamente sempre que seu estado interno for alterado. 
* A re-renderização é otimizada encontrando a diferença entre a interface do usuário do widget antiga e nova e renderizando apenas as alterações necessárias
