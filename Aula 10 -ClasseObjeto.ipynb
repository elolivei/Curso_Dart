## Classe x Objeto

Uma **classe** pode ser considerada um modelo para a criação de objetos.

Vários **objetos** podem ser criados a partir de
uma única classe, o que diferencia os objetos são suas características.

```python
class Employee {
   String name ="";
   
   //getter method
   String get emp_name {
      return name;
   }
   //setter method
   void set emp_name(String name) {
      this.name = name;
   }
   //function definition
   void result() {
      print(name);
   }
}

void main() {
   //object creation
   Employee emp = new Employee();
   emp.name = "employee1";
   emp.result(); //function call
}
```

**Nomes de classe** 

* Deve começar com a letra maiúscula.
* Deve ser um substantivo como Cor, Botão, Sistema, Tópico, etc.
* Use palavras apropriadas, em vez de siglas.

**Nomes das instâncias (objetos)** 

**Nomes de Variável/Atributo**

* Ele deve começar com uma letra minúscula, como id, nome.
* Não deve começar com caracteres especiais como & (e comercial), $ (dólar), _ (sublinhado).
 *Se o nome contiver várias palavras, inicie-o com a letra minúscula seguida por uma letra maiúscula como firstName, lastName.
* Evite usar variáveis de um caractere, como x, y, z.


### **Atributos:** 

Os atributos dizem respeito as características do objeto. No mundo real, qualquer objeto possui elementos que o caracterizam, na linguagem de programação, essas características são mapeadas pelos atributos.

```python
class Produto{
	// declaracao de atributos
	String descricao = "";
	double preco = 0.0;
}

void main(){
    
// Cria um novo produto 	
Produto mochila = new Produto();

//atributos recebendo os seus valores / referências
mochila.descricao = "Mochila grande";
mochila.preco = 50.0;

print("Descricao: " + mochila.descricao);
print("Preco: " + mochila.preco.toString());

Produto tv = new Produto();

//atributos recebendo os seus valores / referências
tv.descricao = "TV 50 polegadas";
tv.preco = 2999.0;

//tv = mochila; // o que acontece se descomentar esta linha

print("Descricao: " + tv.descricao);
print("Preco: ${tv.preco}"); 

}
```

### **Prática**
- Crie uma classe Pessoa:
  - Crie dois atributos, nome do tipo String e idade do tipo int.
- No método main:
   - configure os atributos nome e idade dos dois objetos criados e imprima seus valores (print).



### Métodos

Ações ou eventos, na linguagem de programação, são
modelados pelos métodos.

**Nomenclatura Método**

* Deve começar com letra minúscula.
* Deve ser um verbo como main(), print(), println()
* Se o nome contiver várias palavras, inicie-o com uma letra minúscula seguida por uma letra maiúscula, como actionPerformed().


```python
class Produto{
	String descricao = "";
	double preco = 0.0;
  
	// tipoRetorno nomeMetodo (lista de parametros)
	// 	(tipoParametro nomeParametro, tipoParametro nomeParametro)

	void inicializaProduto(String novaDescricao, double novoPreco){
		descricao = novaDescricao;
		preco = novoPreco;
	}

	void imprime(){
		print("Descricao: $descricao custa $preco em reais");
	}

}

void main(){
  // o uso de new para criação de objeto é opcional
  Produto mochila = Produto();
  mochila.inicializaProduto("Mochila grande",50.0);
  mochila.imprime();

  Produto tv = Produto();
  tv.inicializaProduto("TV 50 polegadas",2999.0);
  //tv = mochila; // o que acontece se descomentar esta linha
  tv.imprime();
		
}
```

### **Prática**
- Na classe Pessoa:
  - Crie um método inicializaPessoa que receba dois parâmetros e inicialize os dois atributos do objeto (nome e idade).
  - Crie um método imprime que imprima os atributos.
- Na função main:
  - Faça uso dos dois métodos para inicializar e imprimir os atributos dos objetos criados.
