## Declarações if-else em Dart 
A declaração if do Dart é usada para testar uma condição. Verifica condições boolean: true ou false. Existem vários tipos de declarações if em Dart.

* Declaração if
* Declaração if-else
* Declaração if-else-if encadeado
* Aninhamento de declarações if


### Declaração if em Dart
image.png

```python
    //Programa Dart para demonstrar o uso da declaração if  

import 'dart:math';

void main(){  
  int rand = Random().nextInt(10);      
    
  if(rand > 5){
    print("$rand é maior que 5");
  }
print("Após o if"); 
}  
```

###  Declaração if-else em Dart

A declaração if-else em Dart também testa uma condição. Executa se a condição do bloco if é verdadeira caso contrário o bloco else é executado.<br>
image.png

```python
    //Um programa Dart para demonstrar o uso da declaração if-else.  
    //Este é um programa que verifica números pares e impares  
import 'dart:math';

void main() {  
				 	
	print("\tTestando se é par");
	int numero = Random().nextInt(10);
	
	bool ePar = numero % 2 == 0;

	if(ePar){
		print("Numero $numero  e par");
	}else{
		print("Numero $numero nao e par");
	}     
		
}  
```

### Declaração if-else-if encadeado em Dart

A declaração if-else-if encadeada executa uma condição para múltiplas declarações.
image.png

```python
import 'dart:math';

void main(){
		
	int numero = Random().nextInt(10);
	print("\tTestando se é maior que 5");

	if(numero>5){
		print("Numero $numero e maior que 5");
	}else if(numero < 5){
		print("Numero $numero e menor que 5");
	}else{
		print("Numero $numero e igual a 5");
	}

}
```

###  Declaração Switch no Dart

A declaração switch executa uma declaração vinda de múltiplas condições. É como o if-else-if.
image.png

```python
import 'dart:math';

void main(){
	int mes = Random().nextInt(12);
	switch(mes){
		// case (label) ponto de entrada
		case 4:
		case 6:
		case 9:
		case 11:
			print("mes $mes Tem 30 dias");
			break; // sai do bloco
		case 2:
			print("mes $mes Tem 28 ou 29 dias");
			break;
		default: 
			print("mes $mes Tem 31 dias");
			break;			
	}
} 

```
