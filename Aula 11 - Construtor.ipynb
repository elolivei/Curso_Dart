#Construtores
O (pseudo-)método construtor determina que ações devem ser executadas quando da criação de um objeto.

No momento em que um construtor é invocado, a seguinte seqüência de ações é executada para a criação de um objeto:



1.   O espaço para o objeto é alocado.

2. Os membros da classe são inicializados para o objeto, seguindo a ordem em que foram declarados na classe.


### O que acontece quando compilamos uma classe na qual não definimos um construtor?

Quando nao e definido nenhum construtor o compilador cria um construtor padrao (sem parametro) e sem corpo para a classe public Produto(){}

#Exemplo construtor padrão
No exemplo abaixo podemos identificar usamos o construtor padrão do Dart e observamos o código de inicialização sendo executado.

```python
class Produto{
	String descricao = "";
	double preco = 0.0;
	double precoMinimo = 0.0;

	//# Quando nao e definido nenhum construtor
	//# o compilador cria um construtor padrao (sem parametro) para a classe public Produto(){}

	
	void imprime(){
		print("Descricao: $descricao custa $preco em reais");
	}
}

main(){
  Produto mochila = Produto();
  mochila.imprime();

  Produto tv = Produto();
  tv.imprime();

}
```

###Qual é a sintaxe para definição de um construtor? E para invocar (chamar) um construtor?
-Na linguagem Dart, o termo construtor é **semelhante** aos métodos que executamos na programação, exceto que as regras abaixo devem ser obedecidas:

1. O nome do construtor deve ser sempre igual ao nome da classe.
2. Não deve haver nenhum tipo de retorno do construtor.
3. Pode haver um construtor parametrizado e um construtor sem argumento (construtor padrão) em uma classe.
4. O construtor sempre é chamado quando um objeto é instanciado.
5. O construtor não deve prosseguir com palavras-chave como abstract, static, final e synchronized.

Vamos ver mais um exemplo:

```python
class Produto{
	String descricao = "";
	double preco = 0.0;
	double precoMinimo = 0.0;

	//# Quando nao e definido nenhum construtor
	//# o compilador cria um construtor padrao (sem parametro) para a classe public Produto(){}

	
	void imprime(){
		print("Descricao: $descricao custa $preco em reais");
	}
}

main() {
  Produto mochila = Produto();
  mochila.imprime();

  Produto tv = Produto();
  tv.imprime();

}
```

## É possível realizar a sobrecarga de um construtor?
- Como em qualquer outro metodo é possivel fazer a sobrecarga dos construtor.


```python
class Produto{
	String descricao = "";
	double preco = 0.0;
	double precoMinimo = 0.0;

//#sobrecarga de construtores parametros diferentes
  
	Produto(this.descricao){
		print("Foi utilizado o construtor com parametro único descrição");
	}
  
	Produto.m(this.descricao, this.preco){
		print("Foi utilizado o construtor com parametros descriçao e preço");
	}
  
	void imprime(){
		print("Descricao: $descricao custa $preco em reais");
	}
}

main(){
  Produto mochila = Produto("Mochila");
  mochila.imprime();

  Produto tv = Produto.m("TV", 1200.0);
  tv.imprime();

}
```
