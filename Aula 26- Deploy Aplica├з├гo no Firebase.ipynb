# Vamos ver como gerar um release Android

### Construa um APK
* Embora os pacotes de aplicativos sejam preferidos em relação aos APKs, há lojas que ainda não suportam pacotes de aplicativos. 
* Nesse caso, crie uma versão APK para cada ABI (Application Binary Interface) de destino.
* Se você concluiu as etapas de assinatura, o APK será assinado. 
* Neste ponto, você pode considerar ofuscar seu código Dart para dificultar a engenharia reversa. 
* Ofuscar Seu código envolve a adição de alguns sinalizadores ao comando Build.

- Na linha de comando
- Entrar no diretório do projeto e Executar: 

flutter build apk --split-per-abi

- Este comando resulta em três arquivos APK:

    - [project]/build/app/outputs/apk/release/app-armeabi-v7a-release.apk
    - [project]/build/app/outputs/apk/release/app-arm64-v8a-release.apk
    - [project]/build/app/outputs/apk/release/app-x86_64-release.apk

* Removendo o sinalizador resulta em um APK gordo que contém seu código compilado para todas as ABIs de destino. Tais APKs são maiores em tamanho do que suas contrapartes divididas, fazendo com que o usuário faça o download binários nativos que não são aplicáveis à arquitetura do dispositivo.

- Instalar um APK em um dispositivo
- Siga estas etapas para instalar o APK em um dispositivo Android conectado.

Na linha de comando:

- Conecte seu dispositivo Android ao computador com um cabo USB.
