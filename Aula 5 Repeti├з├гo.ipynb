## Loops (Repetições) em Dart

Em linguagens de programação, loops são usados para executar uma sequencia de instruções / funções repetidamente quando algumas condições tornam-se true. Existem três tipos de loops em Dart.

* Loop for
* Loop while
* Loop do-while


### Simples loop for

Um simples loop for é o que utilizado no C / C++. Nós podemos inicializar a variável, checar a condição e incrementar ou decrementar o valor.
<br>
image.png

```python
void main(){

	print("for contador");
	int inicio = 1;
	int fim = 10;		
	for(int contador = inicio;contador<=fim;contador++){
		print(contador);
	}
			//          i 0 1 2 3 4
	print("for controlanndo indice de array");
	List<int> nums = [2,4,6,8,10];
	for(int i=0; i<nums.length ;i++){
		print("nums[$i]:${nums[i]}");
	}
}
```

### Loop for-each em Dart

O loop for-each é usado para atravessar arrays ou coleções em Dart. 

É mais fácil de usar do que o loop for simples porque não precisamos aumentar o valor e usa notação subscrita.



```python
void main(){

		
	List<int> nums = [for (int i = 1; i <= 10; i+=2) i];

	print("enhanced for / for each");
	
	print("Para cada");
	for (var i in nums) {
		print(i);
	}

	print("Usando uma função lambda");

	nums.forEach((i) {
		print(i);
	});

}
```

###  Loop while em Dart

O loop while em Dart é usado para percorrer uma parte do programa algumas vezes. 

Se o número de iterações não é fixado, é recomendado o uso do loop while.<br>
image.png


```python
void main(){

	int contador = 0;
	int contarAte = 11;
	while(contador < contarAte){
		//pre incremento, teste por contador++
		print(contador++);
	}
}
```

###  Loop do while em Dart

O loop do while do Dart é usado para repetir uma parte do programa algumas vezes. 

Se o número de iterações não é fixo e você precisa executar o loop pelo menos uma vez. É recomendado o uso do loop do while.

O loop do while é executado pelo menos uma vez porque a condição é checada depois do corpo do loop.<br>
image.png

```python
void main(){
	int contador = 1;
	int contarAte = 5; // teste com 0
	do{
		print(contador++);
	}while(contador < contarAte);		
}
```
