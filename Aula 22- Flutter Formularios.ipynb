https://www.alura.com.br/artigos/criando-formulario-com-flutter

### Desafio
* Construa uma aplicação que leia as notas e pesos das notas e imprima um boletim
* O formulario deve pedir Descrição, peso e nota
* A soma dos pesos não podem superar 10
* Os pesos devem estar entre 0 e 10
* As notas devem estar entre 0 e 10
* Quando a soma de pesos for 10 calcula e exibe a nota final e a formula de calculo.
